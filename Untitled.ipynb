```python
%matplotlib inline
from sklearn.neighbors import KernelDensity
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# values = np.random.normal(size=(100, 2))
values = np.concatenate((np.random.rand(100000, 2) * 1000, np.random.rand(25000,2)*50))

x_lim = [np.min(values[:,0]), np.max(values[:,0])]
y_lim = [np.min(values[:,1]), np.max(values[:,1])]
```

```python
sns.distplot(values[:,0])
sns.distplot(values[:,1])
```

```python
plt.plot(values[:,0], values[:,1], 'o')
```

```python
density = KernelDensity()
density.fit(values)
```

```python
samples = density.sample(10)
scores = density.score_samples(samples)

df = pd.DataFrame({ 'x': samples[:,0], 'y': samples[:,1], 'score': scores})
df
```

```python
x = np.linspace(x_lim[0], x_lim[1], 10000)
y = np.linspace(y_lim[0], y_lim[1], 10000)

xx, yy = np.meshgrid(x, y)
xy = []
for i in range(10000):
    for j in range(10000):
        xy.append(np.array([xx[i,j], yy[i,j]]))
xy = np.array(xy)
```

```python
scores = density.score_samples(xy)

print("max: ", np.max(scores))
print("min: ", np.min(scores))

df = pd.DataFrame({ 'x': xy[:,0], 'y': xy[:,1], 'score': scores})
df
```
