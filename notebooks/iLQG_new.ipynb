```python
import numpy as np
from drl.ilqg import ilqg
from drl.env import TwoLinkArm

env = TwoLinkArm(g=0.)
```

```python
dyn = lambda x, u: env.dynamics_func(x, u)
cst = lambda x, u: env.cost_func(x, u)
T = 5

x = env.reset()
u = np.random.randn(T, env.action_dim)*15.

for i in range(100):
    env.render()
    
    x, u, L, Vx, Vxx, cost = ilqg(dyn, cst, x, u, {})
    
    x, _, _, _ = env.step(u[0,:])
    
    print('Iter %d, Cost: %s' % (i, cost))
    
```

```python
print(x)
```

```python
_, _, xgoal, ygoal = env.to_cartesian(env.goal)
_, _, xcart, ycart = env.to_cartesian(x)

d = np.sqrt((xgoal-xcart)**2 + (ygoal-ycart)**2)

print('goal: [%.2f, %.2f], pos: [%.2f, %.2f], distance: %.2f' % (xgoal, ygoal, xcart, ycart, d))
```
