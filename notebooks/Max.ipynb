```python
import numpy as np
import pandas as pd

filename = 'data.txt'

# Dit is een dummy data set
data = np.random.randn(100, 4)

pd.DataFrame({'a': data[:,0], 'b': data[:,1], 'c': data[:,2], 'd': data[:,3]})
```

## Opslaan in tekst bestand

```python
f = open(filename, 'w') 

# je heb 3 argumenten om een file te openen: r (read), w (write) and a (append). 
# Let open met het gebruiken van w! De huidige inhoud van een bestand wordt gewist als je daarmee een tekst bestand opent.

for row in range(data.shape[0]):
    # Je schrijft data naar een bestand per regel
    data_string = '' 
    for column in range(data.shape[1]):
        data_string += str(data[row, column]) + ';' # Voeg ieder data punt toe
#         if column < Data.shape[1] - 1:
#             data_string += ';' # Voeg een scheiding tussen de waarde toe, zorg dat dit een teken is dat niet in de waarde zelf kan voorkomen
    data_string += '\n' # Begin nieuwe regel, anders komen alle waarden achter elkaar te staan
    f.write(data_string)
    
f.close()
```

## Bestand lezen

```python
f = open(filename, 'r')

data_read = []
for line in f:
    data_array = line.split(';')[:-1] # '\n' staat achter een ; en wordt ook nog toegevoegd, met [:-1] pakken we alle waarde tot de voorlaatste 
    row = np.zeros(len(data_array))
    for i in range(len(data_array)):
        row[i] = float(data_array[i])
    data_read.append(row)

data_read = np.array(data_read)
    
pd.DataFrame({'a': data_read[:,0], 'b': data_read[:,1], 'c': data_read[:,2], 'd': data_read[:,3]})
```
