```python
import numpy as np
from sklearn.neighbors import KernelDensity
from scipy.ndimage import gaussian_filter
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.rand(100, 1)

sns.kdeplot(x[:,0])
```

```python
kd = KernelDensity(bandwidth=0.15)
kd.fit(x)

test_x = np.sort(x, axis=0)
scores = kd.score_samples(test_x)

plt.plot(test_x, scores)
```

```python
img = np.zeros(100)
inds = x + np.min(x)
inds = (inds/np.max(inds) * 99).astype('uint')
print(inds[:,0])
for i in range(inds.shape[0]):
    img[inds[i]] += 1
print(img)
img = gaussian_filter(img, 10)
print(img)
plt.plot(img)
    
```
